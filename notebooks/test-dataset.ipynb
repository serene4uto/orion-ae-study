```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming your array is called 'signals' with shape (4, 1, signal_len)
signals = dataset[30302]['final']  # or your array

# Remove the middle dimension: (4, 1, signal_len) -> (4, signal_len)
signals_2d = signals.reshape(4, -1)

# Create x-axis (time/sample indices)
x = np.arange(signals_2d.shape[1])

# Plot all 4 signals on the same axes
plt.figure(figsize=(12, 6))
for i in range(4):
    plt.plot(x, signals_2d[i], label=f'Signal {i+1}')

plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.title('4 Signals Overlaid')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
